```python
from tkinter import *

root = Tk()
root.title("Distance Between 2 Points")
root.geometry("744x134")
root.minsize(744,334)
root.maxsize(744,334)
dist = StringVar()
temp=0

def distance():
     
    x1 = float(E1.get())
    y1 = float(E2.get())
    x2 = float(E3.get())
    y2 = float(E4.get())
    temp=((x1-x2)**2 + (y1-y2)**2)**(1/2)
    dist.set(str(temp))


L1 = Label(root, text="Enter the value of A,B and C in comparsion to the general form: Ax^2 + Bx+C = 0", font="helvetica 10 bold", fg="red",padx=10,pady=10)
L1.grid(row=0, column=0, columnspan=2)
Lx1 = Label(root, text="X1 = ")
Lx1.grid(row=1, column=0)
Ly1 = Label(root, text="Y1 = ")
Ly1.grid(row=1, column=2)
Lx2 = Label(root, text="X2 = ")
Lx2.grid(row=2, column=0)
Ly2 = Label(root, text="X2 = ")
Ly2.grid(row=2, column=2)
E1 = Entry(root)
E1.grid(row=1, column=1)
E2 = Entry(root)
E2.grid(row=1, column=3)
E3 = Entry(root)
E3.grid(row=2, column=1)
E4 = Entry(root)
E4.grid(row=2, column=3)
L4 = Label(root, text="Distance = ")
L4.grid(row=5, column=0)
E5 = Entry(root, textvariable=dist)
E5.grid(row=5, column=1)
Button(root, text="Find", fg="red", command=lambda: distance()).grid(row=4, column=1)
mainloop()
```
